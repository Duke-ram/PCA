```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data=sb.load_dataset('iris')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
data['species']=lb.fit_transform(data.species)
```

```python
data
```

```python
from sklearn import model_selection
from sklearn.model_selection import train_test_split
```

```python
x_ind=data.drop(['species'],axis=1)
```

```python
x_ind
```

```python
y_dep=data.species
```

```python
y_dep
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_ind,y_dep,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaler=scaler.fit_transform(x_ind)
```

```python
x_scaler
```

```python
from sklearn.decomposition import PCA
pca=PCA()
x_new=pca.fit_transform(x_scaler)
```

```python
x_new
```

```python
import numpy as np
cov_mat=np.cov(x_scaler.T)
eigen_vals,eigen_vecs=np.linalg.eig(cov_mat)
print('\nEIGENVECTOR \n',eigen_vecs)
```

```python
print('\n EIGENVALUES \n',eigen_vals)
```

```python
exp_variance=pca.explained_variance_ratio_
```

```python
exp_variance
```

```python
plt.figure(figsize=(6,4))
plt.bar(range(4),exp_variance)
```

```python
pca_new=PCA(n_components=2)
x_new1=pca_new.fit_transform(x_scaler)
```
